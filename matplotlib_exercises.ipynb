```python
import matplotlib.pyplot as plt
import math
```

## Q1

```python
x = range(-100, 100)
y = [(num**2 - num + 2) for num in x]
```

```python
plt.plot(x,y)

plt.title('$y = x^2 - x +2$')

plt.xlabel('$x$')
plt.ylabel('$y$')

plt.annotate('Origin', xy = (0,0), xytext = (0, 0))

plt.ylim(0,10000)

#plt.savefig('figure1.png')
```

## Q2

```python
x1 = range(0,100)
y1 = [math.sqrt(n) for n in x1]

plt.plot(x1,y1)

plt.title('$ y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$\sqrt{x}$')
plt.xlim(0,100)
plt.ylim(0, 10)

plt.grid(True, ls = '--')
```

```python
x2 = range(0,20)
y2 = [n**3 for n in x2]

plt.plot(x2,y2)

plt.title('$ y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$x^3$')
plt.xticks(range(0,20,5))
plt.xlim(0,20)
plt.ylim(0)
plt.grid(True, ls = '--')
```

```python
x3 = [x * 0.1 for x in range(0,150)]
y3 = [2 ** n for n in x3]

plt.plot(x3,y3)

plt.title('$ y = 2^x$')
plt.xlabel('$x$')
plt.ylabel('$2^x$')
plt.grid(True, ls = '--')
```

```python
x4 = [x * 0.1 for x in range(0,150)]
y4 = [1/(n+1) for n in x4]

plt.plot(x4,y4)

plt.title('$ y = 1/(x+1)$')

plt.xlabel('$x$')
plt.ylabel('$1/(x+1)$')
plt.grid(True, ls = '--')

plt.show()
```

## Q3

```python
#plot1
plt.figure(figsize = (10,6))
plt.subplot(2,2,1)
plt.plot(x1,y1)

plt.title('$ y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$\sqrt{x}$')
plt.xlim(0,100)
plt.ylim(0, 10)
plt.grid(True, ls = '--')

#plot 2
plt.subplot(222)

plt.plot(x2,y2)

plt.title('$ y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$x^3$')
plt.xticks(range(0,20,5))
plt.xlim(0,20)
plt.ylim(0)
plt.grid(True, ls = '--')


#plot 3
plt.subplot(223)

plt.plot(x3,y3)
plt.grid(True, ls = '--')
plt.xlabel('$x$')
plt.ylabel('$2^x$')
plt.title('$y = 2^x$')

#plot 4
plt.subplot(224)

plt.plot(x4,y4)
plt.title(r'$ y = \frac{1}{x+1}$')
plt.xlabel('$x$')
plt.ylabel(r'$\frac{1}{x+1}$')
plt.grid(True, ls = '--')
plt.xlim(0,15)
plt.ylim(0)


plt.tight_layout()
```

## Q4

```python
plt.figure(figsize = (8,6))

plt.plot(x1, y1, c='red', label='$\sqrt{x}$', alpha=1.0)
plt.plot(x2, y2, c='orange', label='$x^{3}$', alpha=.8)
plt.plot(x3, y3, c= 'blue', label='$2^{x}$', alpha=.8)
plt.plot(x4, y4, c= 'green', label='$1/(x+1)$', alpha=.8)

plt.ylim(0, 10)
plt.xlim(0, 10)

plt.ylabel('$y$')
plt.xlabel('$x$')
plt.title('4 functions')
plt.legend()

plt.show()
```
